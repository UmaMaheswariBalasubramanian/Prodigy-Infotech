# Import necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load  test dataset 

```python
test_data = pd.read_csv('test.csv')
```

# Initial Data Exploration:



### Display the first few rows of the dataset:

```python
print(test_data.head())
```

```python
test_data.head(5)
```

```python
test_data.columns
```


### Get basic statistics of the dataset:

```python
print(test_data.describe())
```

### Check data types and missing values:

```python
print(test_data.info())
```

# Univariate Analysis:

### Histogram for a numeric column (e.g., 'Age'):

```python
plt.hist(test_data['Age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### Bar plot for a categorical column (e.g., 'Sex'):

```python


sns.countplot(test_data['Sex'])
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

# Bivariate Analysis:



### Scatter plot to explore the relationship between two numeric columns (e.g., 'Age' vs. 'Fare'):

```python
plt.scatter(test_data['Age'], test_data['Fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

### Box plot to visualize the distribution of a numeric column by a categorical variable (e.g., 'Gender' vs. 'Income'):

```python
sns.boxplot(x='Sex', y='Fare', data=test_data)
plt.xlabel('Sex')
plt.ylabel('Fare')
plt.show()
```

# Multivariate Analysis:

### Correlation matrix to identify relationships between numeric columns:

```python
correlation_matrix = test_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

# Additional Analysis:



### Pairplot to visualize relationships between multiple numeric columns:

```python
sns.pairplot(test_data, diag_kind='kde')
plt.show()
```
